# MODEL 1:

**Importing Libraries**

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

**Reading CSV file**

```python
data = pd.read_csv("D:\coding\ML_Assignments/temperatures.csv")
```

```python
data
```

**Splitting DataSet**

```python
from sklearn.model_selection import train_test_split
x1 = data[["YEAR"]]
y1 = data[["JUN"]]
X_train1 , X_test1, Y_train1 , Y_test1 = train_test_split(x1,y1,test_size=0.3)
```

```python
X_train1
```

```python
Y_train1
```

**Importing Linear Model from SKlearn**

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
model = reg.fit(X_train1,Y_train1)
y_predict1 = model.predict(X_test1)
```

```python
y_predict1
```

**Visualization**

```python
import matplotlib.pyplot as plt
plt.scatter(X_train1,Y_train1, color="blue")
plt.plot(X_test1.squeeze(), y_predict1, color="red")
plt.show()
```

**Calculating MAE, MSE, R2**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
print("MSE", mean_squared_error(Y_test1, y_predict1))
print("MAE", mean_absolute_error(Y_test1, y_predict1))
print("R2 Score ", r2_score(Y_test1, y_predict1))
```
