**Dropping unwanted Column**

```python
df.drop(columns=['Serial No.'])
```

```python
nf=df['Chance of Admit ']
```

```python
nf
```

**Applying Logic to make COA 1 when value is >0.8 and else COA 0**

```python
for i in nf:
    if (i>0.8):
        df['Chance of Admit ']=df['Chance of Admit '].replace({i:'1'})
    else:
        df['Chance of Admit ']=df['Chance of Admit '].replace({i:'0'})
```

```python
df
```

**Creating independent Variable**

```python
x=df[["GRE Score","TOEFL Score","University Rating","SOP","LOR ","CGPA"]]
```

```python
x
```

**Creating Dependent Variable**

```python
 y=df[["Chance of Admit "]]
```

```python
y
```

**Splitting Training and Testing Datasets**

```python
from sklearn.model_selection import train_test_split
```

```python
 x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

**Building Decision Tree Model**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier().fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

**Evaluation of Model**

```python
 from sklearn.metrics import confusion_matrix
matrix=confusion_matrix(y_test,y_predict)
```

```python
matrix
```

**Evaluating Accuracy of model**

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_predict)
```

```python
 accuracy
```

```python
 from sklearn.metrics import classification_report
```

```python
report=classification_report(y_test,y_predict)
```

```python
 print(report)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt


class_names_list = ['Class_0', 'Class_1', 'Class_2']

plt.figure(figsize=(10, 6))
plot_tree(
    model,
    fontsize=9,
    rounded=True,
    proportion=True,
    class_names=class_names_list,
    max_depth=2,
    filled=True
)
plt.show()
```
