```python
from sklearn.cluster import KMeans
```

```python
x = df.iloc[:,[3, 4]].values
```

```python
x
```

```python
wcss_list=[]
```

```python
import matplotlib.pyplot as mtp
```

```python
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 55)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
mtp.plot(range(1, 11), wcss_list)
mtp.title('The Elbow Method Graph')
mtp.xlabel('Number of clusters(k)')
mtp.ylabel('wcss_list')
mtp.show()
```

```python
 kmeans = KMeans(n_clusters=5, init='k-means++', random_state= 55)
y_pred= kmeans.fit_predict(x)
```

```python
print(y_pred)
```

```python
import matplotlib.pyplot as plt

# Scatter points for different clusters
plt.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1], s=100, c='blue', label='Cluster 1')
plt.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1], s=100, c='yellow', label='Cluster 2')
plt.scatter(x[y_pred == 2, 0], x[y_pred == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(x[y_pred == 3, 0], x[y_pred == 3, 1], s=100, c='red', label='Cluster 4')
plt.scatter(x[y_pred == 4, 0], x[y_pred == 4, 1], s=100, c='orange', label='Cluster 5')

# Scatter the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='black', label='Centroid')

# Set the title and labels
plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```
